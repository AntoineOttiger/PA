```python
import numpy as np
import matplotlib.pyplot as plt

# Exemple : des points en 3D
n_points = 500
x = np.random.uniform(0, 10, n_points)
y = np.random.uniform(0, 10, n_points)
z = np.sin(x) * np.cos(y)  # Ou n'importe quelle valeur

# Option 1 : utiliser 2D histogram
heatmap, xedges, yedges = np.histogram2d(x, y, bins=100, weights=z)

# heatmap.shape => (100, 100)

# Affichage
plt.imshow(
    heatmap.T,     # Transposé pour que x et y soient bien orientés
    origin='lower',
    cmap='viridis'
)
plt.colorbar(label="Valeur moyenne de z")
plt.title("Heatmap 2D des points 3D")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
